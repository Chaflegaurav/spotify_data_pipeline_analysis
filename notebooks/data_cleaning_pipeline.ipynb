```python
names = []
ids = []
duration = []
album_type = []
popularity = []
artists_name = []
for track in tracks['items']:
    names.append(track['track']['name'])
    ids.append(track['track']['id'])
    duration.append(track['track']['duration_ms']/1000)
    album_type.append(track['track']['album']['album_type'])
    popularity.append(track['track']['popularity']) 
    artists = ', '.join([artist['name'] for artist in track['track']['artists']])
    artists_name.append(artists)

        
```

```python
import pandas as pd
df = pd.DataFrame({'name':names, 'id':ids, 'duration': duration, 'album_type': album_type, 'popularity':  popularity, 'artists_name': artists_name})
```

```python
# Check for missing values
print("\n Missing values before cleaning:")
print(df.isnull().sum())
```

```python
# Remove duplicates based on track ID
df_cleaned = df.drop_duplicates(subset=['id'], keep='first')
print(f"\n duplicates removed: {len(df) - len(df_cleaned)}")
```

```python
# Handle Missing values
df_cleaned['name'] = df_cleaned['name'].fillna('Unknown Track')
df_cleaned['artists_name']=df_cleaned['artists_name'].fillna('Unknown artists')
df_cleaned['popularity']=df_cleaned['popularity'].fillna(0)
```

```python
# Data type conversions
df_cleaned['duration'] = df_cleaned['duration'].astype(float)
df_cleaned['popularity'] = df_cleaned['popularity'].astype(int)
df_cleaned
```

```python
# Add calculated columns
df_cleaned['duration_minutes'] = (df_cleaned['duration'] / 60).round(2)
df_cleaned
```

```python
# Categorized popularity
def categorize_popularity(score):
    if score >= 80:
         return 'Very High'
    elif score >= 60:
          return 'High'
    elif score >= 40:
         return 'Medium'
    elif score >= 30:
         return 'Low'
    else:
          'Very Low'

df_cleaned['popularity_category'] = df_cleaned['popularity'].apply(categorize_popularity)
    
df_cleaned
```

```python
# Categorize duration
def categorize_duration(seconds):
    minutes = seconds / 60
    if minutes < 2.5:
        return 'Short'
    elif minutes < 4:
        return 'Medium'
    else:
          'Long'

df_cleaned['duration_category'] = df_cleaned['duration'].apply(categorize_duration)
df_cleaned

print("\nData cleaning complete!")
print("\nCleaned DataFrame Info:")
print(df_cleaned.info())

print("\nSummary Statistics:")
print(df_cleaned[['duration', 'popularity']].describe())

print("\nAlbum Type Distribution:")
print(df_cleaned['album_type'].value_counts())

print("\nPopularity Category Distribution:")
print(df_cleaned['popularity_category'].value_counts())
```
